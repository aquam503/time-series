Réalisation du test augmenté de Dickey-Fuller pour vérifier la stationnarité pour une série temporelle non saisonnière :

*Note*

Il existe deux types de test de stationnarité différents : 

* Les tests standards pour lesquels l'hypothèse H0 concerne la présence de racine unitaire contre l'hypthèse H1 de stationnarité du processus. On peut citer parmi ces derniers le test de Dickey-Fuller(1979), le test augmenté de Dickey-Fuller(1981) ou encore le test non paramétrique de Phillips-Perron (1988) et le test de Seo (1999).

* Les tests de stationnarité pour lesquels l'hypothèse nulle est que la série est stationnaire. Et donc l'hpothese H1 est la non stationnarité. Parmi ces tests, on peut citer comme le test KPSS(1992), les tests de Leybourne et Mc Cabe (1994) ou encore celui de Paparoditis (2003).

Ci desouss nous utilisons le test augmenté de Dickey Fuller qui permet de meetre en évidence le caractère stationnaire ou non d'une série temporelle par la détermination d'une tendance déterministe ou stochastique.

```python
from statsmodels.tsa.stattools import adfuller
testadf2 = adfuller(ma1_data)
if (testadf2[1]>0.05):
    print("pvalue =",testadf2[1],"supérieure à 5% la série est donc non stationnaire")
else:
    print("pvalue =",testadf2[1],"La série est donc stationnaire")
```
